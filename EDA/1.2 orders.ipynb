```python
def replace_nulls(df, col_origen, col_destino):
    df[col_destino] = df[col_destino].fillna(df[col_origen].map(lambda x: x))
    return df
```

```python
replace_nulls(df, "order_purchase_timestamp", "order_approved_at")
```

```python
replace_nulls(df, "order_approved_at", "order_delivered_carrier_date")
```

### **`Date's column changes`**

```python
purchase_timestamp = pd.to_datetime(df["order_purchase_timestamp"])
```

```python
approved_at = pd.to_datetime(df["order_approved_at"])
```

```python
delivered_carrier = pd.to_datetime(df["order_delivered_carrier_date"])
```

```python
delivered_customer = pd.to_datetime(df["order_delivered_customer_date"])
```

```python
estimated_delivery = pd.to_datetime(df["order_estimated_delivery_date"])
```

### **`Dataframe with only dates, to find differences (later date - previous date)`**

```python
dates = pd.concat([purchase_timestamp, approved_at, delivered_carrier, delivered_customer, estimated_delivery], axis=1)
```

```python
dates.info()
```

```python
dates["approved_and_purchased_difference"] = (dates["order_approved_at"]-dates["order_purchase_timestamp"])/pd.Timedelta(days=1)
```

```python
dates[dates["approved_and_purchased_difference"]<0]["approved_and_purchased_difference"].count()
```

```python
dates["approved_and_carrier_difference"] = (dates["order_delivered_carrier_date"] - dates["order_approved_at"])/pd.Timedelta(days=1)
```

```python
dates[dates["approved_and_carrier_difference"]<0]["approved_and_carrier_difference"].count()
```

```python
# Max difference
dates[dates["approved_and_carrier_difference"]<0]["approved_and_carrier_difference"].min()
```

```python
dates[dates["approved_and_carrier_difference"]<0]["approved_and_carrier_difference"].idxmin()
```

```python
# Min difference
dates[dates["approved_and_carrier_difference"]<0]["approved_and_carrier_difference"].max()
```

```python
dates["client_and_carrier_difference"] = (dates["order_delivered_customer_date"] - dates["order_delivered_carrier_date"])/pd.Timedelta(days=1)
```

```python
dates[dates["client_and_carrier_difference"]<0]["client_and_carrier_difference"].count()
```

```python
dates["client_and_estimated_delivery_difference"] = (dates["order_estimated_delivery_date"] - dates["order_delivered_customer_date"])/pd.Timedelta(days=1)
```

**Late deliveries:**
