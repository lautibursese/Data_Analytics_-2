```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
our_data = os.listdir("C:/Users/Predator/Desktop/Lauti/ProgramaciÃ³n/GitHub/2. Data Analytics/#2/datasets")
our_data
```

```python
data_dict = {}
for file in our_data:
    
    path = f"C:/Users/Predator/Desktop/Lauti/ProgramaciÃ³n/GitHub/2. Data Analytics/#2/datasets/{file}"
    df = pd.read_csv(path)
    data_nan = df.isna().sum().sum()
    rows, columns = df.shape
    percent_nan = data_nan/(rows*columns)
    name_columns = df.columns

    # Get known if at least one culumn is a date-format
    has_date = any(df.columns.str.contains("Date", case=False))

    duplicates = df.duplicated().sum()/rows

    data_dict[file] = {"percent_nan": percent_nan, 
                        "rows": rows, 
                        "columns": columns,
                        "has_date_column": has_date,
                        "percent_duplicates": duplicates}

data_dict
```

```python
data = pd.DataFrame(data_dict.values(), index=data_dict.keys())
data
```

```python
total_data = (data.rows * data["columns"]).sum()
total_data
```

```python
total_nan = (data.percent_nan * data.rows * data["columns"]).sum()
total_nan
```

```python
percent_total_nan = total_nan/total_data
percent_total_nan
```

```python
percent_nan = data.sort_values("percent_nan", ascending=False)
ax = sns.barplot(data=percent_nan, x="percent_nan", y=percent_nan.index)
ax.set(xlabel="Percentage of nulls", ylabel="Datasets")
ax.set_title("Percentage of nulls per dataset")
```

```python
labels = ["Nulls", "Not nulls"]
colors = ["tab:orange", "tab:blue" ]
plt.pie([percent_total_nan, 1-percent_total_nan], labels=labels, colors=colors, autopct='%.0f%%')
plt.title("Total percentage of nulls on datasets")
```
